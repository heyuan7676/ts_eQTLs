```python
### Fig 2 - examples

[Y, W, pV, B] = readin_YW_BP(inputprefix, Nrows_to_read=None)


def get_ywpb(example_pairs, group, suffix):
    Y_shared, W_shared, PV_shared, B_shared = [], [], [] , []

    for p in example_pairs:
        [gene, snp] = p.split(':')
        Y_shared.append(np.array(Y.loc[gene].loc[snp]))
        W_shared.append(np.array(W.loc[gene].loc[snp]))
        PV_shared.append(np.array(pV.loc[gene].loc[snp]))
        B_shared.append(np.array(B.loc[gene].loc[snp]))

    Y_shared = pd.DataFrame(np.array(Y_shared), index=example_pairs)
    W_shared = pd.DataFrame(np.array(W_shared), index = example_pairs)
    PV_shared = pd.DataFrame(np.array(PV_shared), index = example_pairs)
    B_shared = pd.DataFrame(np.array(B_shared), index = example_pairs)
    
    Y_shared[np.isnan(Y_shared)] = 0
    W_shared[np.isnan(W_shared)] = 0
    
    
    Y_shared.to_csv('%s/examples/Fig2_example_group%d_Y_%s.txt' % (savedir, group, suffix), sep='\t')
    W_shared.to_csv('%s/examples/Fig2_example_group%d_W_%s.txt' % (savedir, group, suffix), sep='\t')
    PV_shared.to_csv('%s/examples/Fig2_example_group%d_PV_%s.txt' % (savedir, group, suffix), sep='\t')
    B_shared.to_csv('%s/examples/Fig2_example_group%d_B_%s.txt' % (savedir, group, suffix), sep='\t')

```

```python
def get_ywpb(example_pairs, tis, suffix):
    Y_shared, W_shared, PV_shared, B_shared = [], [], [] , []

    for p in example_pairs:
        [gene, snp] = p.split(':')
        Y_shared.append(np.array(Y.loc[gene].loc[snp]))
        W_shared.append(np.array(W.loc[gene].loc[snp]))
        PV_shared.append(np.array(pV.loc[gene].loc[snp]))
        B_shared.append(np.array(B.loc[gene].loc[snp]))

    Y_shared = pd.DataFrame(np.array(Y_shared), index=example_pairs)
    W_shared = pd.DataFrame(np.array(W_shared), index = example_pairs)
    PV_shared = pd.DataFrame(np.array(PV_shared), index = example_pairs)
    B_shared = pd.DataFrame(np.array(B_shared), index = example_pairs)
    
    Y_shared[np.isnan(Y_shared)] = 0
    W_shared[np.isnan(W_shared)] = 0
    
    
    Y_shared.to_csv('%s/examples/Fig2_example_%s_Y_%s.txt' % (savedir, tis, suffix), sep='\t')
    W_shared.to_csv('%s/examples/Fig2_example_%s_W_%s.txt' % (savedir, tis, suffix), sep='\t')
    PV_shared.to_csv('%s/examples/Fig2_example_%s_PV_%s.txt' % (savedir, tis, suffix), sep='\t')
    B_shared.to_csv('%s/examples/Fig2_example_%s_B_%s.txt' % (savedir, tis, suffix), sep='\t')





for tis in ciseQTL_not_in_f.keys():
    print(tis)
    if len(ciseQTL_not_in_f[tis]) == 0:
        continue
    example_pairs = np.random.choice(ciseQTL_not_in_f[tis], 1000, replace = False)
    get_ywpb(example_pairs, tis, 'cisNotinFactor')
```

```python
### enrichment of chromHMM
enrich_dir = '/work-zfs/abattle4/heyuan/tissue_spec_eQTL_v8/downstream/enrichmentTest/active_regions'
#LMfn = 'flashr_Loadings_beta_BH_alpha0.05_corrected'
LMfn = '%s%s_Loadings_projection' % (FMfn, LDprefix)

rep_chromm = ['10_TssBiv', '11_BivFlnk', '12_EnhBiv', '13_ReprPC', '14_ReprPCWk','9_Het', '15_Quies']
act_chromm = ['1_TssA', '2_TssAFlnk', '3_TxFlnk', '4_Tx', '5_TxWk', '6_EnhG','7_Enh', '8_ZNF_Rpts']


def plot_to_background(background, group):
    
    df = pd.DataFrame(columns=['Oddsratio', 'pvalue', 'group', 'inTissue', 'matched', 'element'])
    
    ## chromHMM
    for sub in rep_chromm + act_chromm:
        try:
            rs = pd.read_csv('%s/%s_Active_enrichment_ROADMAP_%s_to%s.txt' %(enrich_dir, LMfn, sub, background), sep=',')
            rs = rs[rs['group'] == group]
            rs['element'] = [sub] * len(rs)
            df = pd.concat((df, rs), sort=True)
        except:
            print([group, sub])

    ## DNase
    try:
        dnase = pd.read_csv('%s/%s_Active_enrichment_%s_to%s.txt' %(enrich_dir, LMfn, 'DNase_None', background), sep=',')
        dnase = dnase[dnase['group'] == group]
        dnase['element'] = ['DNase'] * len(dnase)
        df = pd.concat((df, dnase), sort=True)
    except:
        print([group, 'DNase'])
            
    if len(df) == 0:
        return []
    
    df['BH_p'] = multitest.multipletests(df['pvalue'], method='fdr_bh')[1]
    #df = df[df['BH_p'] < 0.05]
    
    med = pd.DataFrame(df.groupby('element')['Oddsratio'].median())
    med.columns = ['med_OR']
    df = df.merge(med, left_on = 'element', right_index = True)
    
    return df



### outlier pairs compared to random background 

background = 'random'

df_tis_save= plot_to_background(background, 0)
for group in range(1, len(Comp_tissues)):
    df_tis_i = plot_to_background(background, group)
    if len(df_tis_i) > 0:
        df_tis_save = df_tis_save.append(df_tis_i)
        
df_tis_save.to_csv('%s/Fig3_chromHMM_random_MF_projection.txt' % savedir, sep='\t', index = False)


### ts pairs compared to shared pairs
if 0:
    background = 'shared'
    df_tis_save= plot_to_background(background, 1)

    for group in range(2, len(Comp_tissues)):
        df_tis_i = plot_to_background(background, group)
        if len(df_tis_i) > 0:
            df_tis_save = df_tis_save.append(df_tis_i)
        
    df_tis_save.to_csv('%s/Fig3_chromHMM_shared.txt' % savedir, sep='\t', index = False)


    
    ### random pairs - ts pairs compared to shared pairs

    background = 'shared_randomdataSet'
    df_tis_save= plot_to_background(background, 1)
    
    for group in range(2, len(Comp_tissues)):
        df_tis_i = plot_to_background(background, group)
        if len(df_tis_i) > 0:
            df_tis_save = df_tis_save.append(df_tis_i)
        
    df_tis_save.to_csv('%s/Fig3_chromHMM_shared_randomPairs.txt' % savedir, sep='\t', index = False)
```
