```python
def filter_TF_TPM(tf_tmp_med, group, tfNames, TPM):
        passed_tfi = []
        
        ## filter out TFs with median TPM < 1 in matched tissues      
        # deal with names incosistency in TF files
        if 0:
            extra_tfi = list(set(tfNames) - set(tf_tmp_med.index))
            realNames_tfi   = [x.split('::') for x in extra_tfi]
            for i in range(len(realNames_tfi)):
                tfi = realNames_tfi[i]
                tfi = np.intersect1d(tfi, tf_tmp_med.index)
                if len(tfi) < len(realNames_tfi[i]):
                    continue
                else:
                    tfs_pass = np.sum(tf_tmp_med.loc[tfi][Comp_tissues[int(group)]]>TPM, axis=1) == len(Comp_tissues[int(group)])
                    if np.sum(tfs_pass) == len(tfi):
                        passed_tfi.append(extra_tfi[i])
                    
        passed_tfi = []
            
        # other consistent names
        tf_inexp_dataset = np.intersect1d(tfNames, tf_tmp_med.index)
        tf_tmp_med = tf_tmp_med.loc[tf_inexp_dataset]
        tpm_above_idx = np.where(np.sum(tf_tmp_med[Comp_tissues[int(group)]]>TPM, axis=1) > len(Comp_tissues[int(group)])/2.0)[0]
        tfs_pass_tpm = np.array(tf_tmp_med.index)[tpm_above_idx]
        
        allPassed_tf = np.unique(list(tfs_pass_tpm) + list(passed_tfi))
        #assert len(tfs_pass_tpm)+ len(passed_tfi) == len(allPassed_tf)

        return allPassed_tf
```

```python
def qqplot_log10PV(pv_vec, plot = False):
    increment = 1.0 / len(pv_vec)
    x_expect = []
    x_ei = 1.0
    for i in range(len(pv_vec)):
        x_expect.append(x_ei)
        x_ei -= increment
        
    x_expect = np.sort(-np.log10(x_expect))
    pv_vec = np.sort(-np.log10(pv_vec))
    
    if plot:
        plt.figure()
        plt.scatter(x_expect, pv_vec)
        xman = np.max([np.max(x_expect), np.max(pv_vec)])
        plt.plot([0, xman], [0,xman])
        plt.show()
        plt.close()
        
    return [x_expect, pv_vec]
```

```python
import scipy.stats as stats


def filter_HIT(all_OR, HIT = 10, HIT_higher = None):
    all_OR = all_OR.copy()
    #all_OR = all_OR[all_OR['ts_TFBS'] > 0]
    all_OR = all_OR[np.array(all_OR['ts_TFBS'] + all_OR['background_TFBS']) > HIT]
    if HIT_higher is not None:
        all_OR = all_OR[np.array(all_OR['ts_TFBS'] + all_OR['background_TFBS']) < HIT_higher * folds]
    #all_OR = all_OR.copy()
    #all_OR = all_OR[all_OR['OR_background'] > 1]
    all_OR['BH_pvalue'] = multitest.multipletests(all_OR['PV_background'], method = 'fdr_bh')[1]
    all_OR['BH_pvalue_2'] = multitest.multipletests(all_OR['pvalue2'], method = 'fdr_bh')[1]
    all_OR = all_OR.sort_values('BH_pvalue')
        
    return all_OR




def wrap_up(all_OR, HIT = 50, alpha = 0.1, title_text = 'NONE'):

    allOR1 = filter_HIT(all_OR, HIT = HIT)
    
    if 1:    
        x  = allOR1['PV_background']
        plt.figure()
        plt.hist(x, bins=50)
        plt.title(title_text, size=20)
        plt.show()
        plt.close()
        
        qqplot_log10PV(x, plot = True)

    dat_sig = allOR1[allOR1['BH_pvalue'] < alpha]
    tf_count = pd.DataFrame.from_dict(Counter(dat_sig['TF']), orient = 'index').sort_values(0, ascending = False)
    tf_count.columns = ['count']
    #tf_count['TF'] = [x.split(':')[0] for x in tf_count.index]
    tf_count['TF'] = tf_count.index
    dat_sig = dat_sig.merge(tf_count, left_on = 'TF', right_on = 'TF').sort_values('BH_pvalue')
    dat_sig_unique = dat_sig[dat_sig['count'] == 1].sort_values(['group', 'BH_pvalue'])

    print('Among %d(%d) tested TFs, %d(%d) are significantly enriched/depleted, including %d TFs exist in only one factor' % (len(allOR1), len(set(allOR1.index)), len(dat_sig), len(set(dat_sig['TF'])), len(dat_sig_unique)))
    return [allOR1, dat_sig, dat_sig_unique]



def plot_results(allOR1, dat_sig, dat_sig_unique, alpha = 0.1):
    
    dat_sig_enh = dat_sig[dat_sig['OR_background'] > 1]
    dat_sig_dep = dat_sig[dat_sig['OR_background'] < 1]

    xticks_loadings = loading_names()
    plt.figure()
    sns.set(rc={'figure.figsize':(12, 6)}, font_scale=2)
    sns.set_style('whitegrid')
    ax = sns.boxplot(x = 'group', y ='OR_background', data=allOR1, showfliers = False)
    ax.set(xlabel='loading', ylabel='')
    plt.xticks(rotation=90)
    plt.title('Distribution of odds ratio for all tested TFs')
    
    
    plt.figure()
    sns.set(rc={'figure.figsize':(12, 6)}, font_scale=2)
    ax = sns.boxplot(x = 'group', y ='OR_background', data=dat_sig)
    ax.set(xlabel='loading', ylabel='')
    plt.xticks(rotation=90)
    plt.title('Distribution of odds ratio for the enriched TFs \n with BH-corrected p-value < %s' % str(alpha))

    plt.figure()
    sns.set(rc={'figure.figsize':(12, 6)}, font_scale=2)
    sns.set_style('whitegrid')
    y_temp =  dat_sig_enh.groupby('group').size()
    for gt in list(set(allOR1['group']) - set(dat_sig_enh['group'])):
        y_temp[gt] = 0
    y_temp = y_temp.sort_index()
    
    x_temp = [xticks_loadings[l] for l in y_temp.index]
    ax = sns.barplot(x_temp, y_temp, color = 'mediumseagreen')
    ax.set(xlabel='Factors', ylabel='')
    #ax.set_ylim(0, 22)
    plt.xticks(rotation=90)
    plt.title('Number of significantly enriched TFs across factors\n FDR = %s' % str(alpha))


    if len(dat_sig_dep) > 0:
        plt.figure()
        sns.set(rc={'figure.figsize':(12, 6)}, font_scale=2)
        sns.set_style('whitegrid')
        ax = sns.barplot(x=[xticks_loadings[l] for l in dat_sig_dep.groupby('group').size().index], y =dat_sig_dep.groupby('group').size(), color = 'skyblue')
        ax.set(xlabel='Factors', ylabel='')
       #ax.set_ylim(0, 22)
        plt.xticks(rotation=90)
        plt.title('Number of significantly depleted TFs across factors\n FDR = %s' % str(alpha))
    
    if len(dat_sig_unique) > 0:
        plt.figure()
        sns.set(rc={'figure.figsize':(12, 6)}, font_scale=2)
        sns.set_style('whitegrid')
        ax = sns.barplot(x=[xticks_loadings[l] for l in dat_sig_unique.groupby('group').size().index], y =dat_sig_unique.groupby('group').size(), color = 'skyblue')
        ax.set(xlabel='Factors', ylabel='')
       #ax.set_ylim(0, 22)
        plt.xticks(rotation=90)
        plt.title('Number of unique significantly enriched TFs across factors\n FDR = %s' % str(alpha))



```

```python
def readin_enrichment(LMfn, featureName, background,randomPattern, HIT = 0, TPM = 2, folds = 10):
    all_OR = pd.DataFrame()
    
    plt.figure()
    xmax = 0
    for loading in [-1] + list(range(len(Comp_tissues))):
        try:
            fntemps = np.sort([x for x in os.listdir(enrich_dir) if ((LMfn in x) and (featureName in x) and ('_%d_' % loading in x) and ("NULL" not in x))])
            fntemps = np.random.choice(fntemps, folds, replace = False)
            #print(fntemps)
            #fntemps = fntemps[:1]
            dat_tfbs = pd.read_csv('%s/%s' % (enrich_dir, fntemps[0]), sep='\t', index_col = 0, usecols=[0,3,4,5,6])
            for fntemp in fntemps[1:]:
                dat_tfbs_tp = pd.read_csv('%s/%s' % (enrich_dir, fntemp), sep='\t', index_col = 0, usecols=[0,3,4,5,6])
                index_tp = np.unique(list(dat_tfbs.index) + list(dat_tfbs_tp.index))
                dat_tfbs = dat_tfbs.loc[index_tp]
                dat_tfbs_tp = dat_tfbs_tp.loc[index_tp]
                
                for col in dat_tfbs.columns:
                    dat_tfbs.loc[np.isnan(dat_tfbs[col]),col] = 0
                    dat_tfbs_tp.loc[np.isnan(dat_tfbs_tp[col]),col] = 0
                dat_tfbs = dat_tfbs + dat_tfbs_tp
                
        except:
            print('TFBS for group %d does not exist ' % loading)
            #print('%s/%s' % (enrich_dir, fntemps))
            continue

            
        dat_tfbs['ts_TFBS']    = [int(x/len(fntemps)) for x in dat_tfbs['ts_TFBS']]
        dat_tfbs['ts_nonTFBS'] = [int(x/len(fntemps)) for x in dat_tfbs['ts_nonTFBS']]
    
        or_pv = dat_tfbs.apply(lambda x: fisher_exact([[x[2], x[3]], [x[0], x[1]]], alternative='greater'), axis=1)
        dat_tfbs['OR_background'] = [x[0] for x in or_pv]
        dat_tfbs['PV_background'] = [x[1] for x in or_pv]
        
        #dat_tfbs.index = [x.split('-')[0].split('::')[0].upper() for x in dat_tfbs.index]
        dat_tfbs.index = [x.upper() for x in dat_tfbs.index]
        left_tfs = filter_TF_TPM(tf_tmp_med, loading, np.unique(dat_tfbs.index), TPM = TPM)
        #left_tfs = np.array(dat_tfbs.index)
        dat_tfbs = dat_tfbs.loc[left_tfs].sort_values('PV_background').copy()
        dat_tfbs = dat_tfbs[np.array(dat_tfbs['ts_TFBS'] + dat_tfbs['background_TFBS']) > HIT]
        dat_tfbs['group'] = [loading] * len(dat_tfbs)    
        all_OR = all_OR.append(dat_tfbs)
            
        print(loading, dat_tfbs.shape)
        
        if 0:
            x = dat_tfbs['PV_background']
            [a, b] = qqplot_log10PV(x)
            xmax = np.max([xmax, np.max(a), np.max(b)])
            plt.scatter(a, b, s=10)
            
    plt.plot([0, xmax], [0, xmax])
    plt.title('qqplot of p-value from chi-square tests across factors')
    plt.show()
    plt.close()

    all_OR['pvalue2'] = all_OR.apply(lambda x: chi2_contingency([[x['ts_TFBS'], x['ts_nonTFBS']], [x['background_TFBS'], x['background_nonTFBS']]])[1], axis=1)
    all_OR['TF'] = all_OR.index
    
    ## GO score
    tfi_score = []       
    for i in range(len(all_OR)):
        rowi = all_OR.iloc[i]
        score_i = score_go(rowi['TF'], Comp_tissues[rowi['group']])
        #score_i = -1
        tfi_score.append(score_i)    
    all_OR['score'] = tfi_score
        
    #for tis_dat in dat.columns:
     #   all_OR['%s_score' % tis_dat] = [score_go(x, [tis_dat]) for x in all_OR['TF']]
    
    all_OR = all_OR.sort_values('PV_background')
    #all_OR.index = all_OR.apply(lambda x: ':'.join((x['TF'], str(x['group']))), axis=1)
    #all_OR = all_OR[~all_OR.index.duplicated()]
    
    x  = all_OR['PV_background']
    plt.figure()
    plt.hist(x, bins=20)
    plt.show()
    plt.close() 
    
    return all_OR


def score_go(gi, ti):
    ts = np.intersect1d(ti, dat.columns)
    if len(ts) == 0:
        return -1
    if gi not in gene_gos.keys():
        return -1
    try:
        #goi = gene_gos[gene_annotation[gi][0]]
        goi = gene_gos[gi]
        maxscore = np.nanmax(np.array(dat.loc[np.array(goi)][ts]).ravel())
    except:
        return 0
    return maxscore


def save_enriched_TFs_forGSEA(df, alpha):
    df = df[df['BH_pvalue'] < alpha]
    tp = df.set_index('group')

    for k in range(23):
        try:
            tfs_group = np.array(tp.loc[k]['TF']).ravel()
        except:
            tfs_group = []
            continue
        if len(tfs_group) == 0:
            continue
        fn = open('../tfs_gsea/tfs_enriched_alpha%s_group%d_torandom.txt' % (str(alpha), k), 'w')
        for i in tfs_group:
            i = i.replace('-VAR.2', '')
            if len(i.split("::")) > 1:
                for ii in i.split("::"):
                    try:
                        fn.write('%s\t%s\n' % (gene_annotation_rev[ii], ii))
                    except:
                        continue
                        #print(ii)
        fn.close()  
        
        
```
