```python
def readin_gsea_result_tis(geneGroup, HIT, prefix = 'Thresholding', lower = 20, higher = 500):
    gsea_dat = pd.DataFrame()

    for tis in tissues:
        gsea_fn = '%s/%s_eGenes_%s_%s_enrichment_%s.txt' % (gsea_dir, prefix, geneGroup, tis, gsea_file_used)
        try:
            gsea_dat_tis = pd.read_csv(gsea_fn, sep='\t', index_col = 0)
            if np.sum(gsea_dat_tis['test_inset']!=0) == 0:
                continue
        except:
            continue

        gsea_dat_tis['group'] = tis
        gsea_dat = gsea_dat.append(gsea_dat_tis)
    
    gsea_dat = gsea_dat[gsea_dat['test_inset'] + gsea_dat['background_inset'] > HIT]
    gsea_dat['N_genes'] = [len(gsea_pathways[g.replace(' ','')]) for g in np.array(gsea_dat.index)]
    gsea_dat = gsea_dat[gsea_dat['N_genes'] < higher]
    gsea_dat = gsea_dat[gsea_dat['N_genes'] > lower]
    
    #gsea_dat['ts_pvalue'] = list(map(lambda x: fisher_exact([[x[0], x[1]], [x[2], x[3]]])[1], np.array(gsea_dat[gsea_dat.columns[:4]])))
    gsea_dat['BH_p'] = multitest.multipletests(gsea_dat['pvalue'], method = 'fdr_bh')[1]
    gsea_dat['Bonf_p'] = multitest.multipletests(gsea_dat['pvalue'])[1]
    gsea_dat = gsea_dat.sort_values('pvalue')
    
    return gsea_dat
```

```python
def readin_gsea_result(geneGroup, HIT, lower = 20, higher = 500):
    gsea_dat = pd.DataFrame()

    for g in range(len(Comp_tissues)):
        gsea_fn = '%s/%s_%s_group%d_enrichment_%s.txt' % (gsea_dir, LMfn,geneGroup, g, gsea_file_used)
        try:
            gsea_dat_tis = pd.read_csv(gsea_fn, sep='\t', index_col = 0)
            if np.sum(gsea_dat_tis['test_inset']!=0) == 0:
                continue
        except:
            continue

        gsea_dat_tis['group'] = g
        gsea_dat = gsea_dat.append(gsea_dat_tis)
    
    gsea_dat = gsea_dat[gsea_dat['test_inset'] + gsea_dat['background_inset'] > HIT]
    gsea_dat['N_genes'] = [len(gsea_pathways[g.replace(' ','')]) for g in np.array(gsea_dat.index)]
    gsea_dat = gsea_dat[gsea_dat['N_genes'] < higher]
    gsea_dat = gsea_dat[gsea_dat['N_genes'] > lower]
    
    #gsea_dat['ts_pvalue'] = list(map(lambda x: fisher_exact([[x[0], x[1]], [x[2], x[3]]])[1], np.array(gsea_dat[gsea_dat.columns[:4]])))
    gsea_dat['BH_p'] = multitest.multipletests(gsea_dat['pvalue'], method = 'fdr_bh')[1]
    gsea_dat['Bonf_p'] = multitest.multipletests(gsea_dat['pvalue'])[1]
    gsea_dat = gsea_dat.sort_values('pvalue')
    

    return gsea_dat
```

```python
### only for bp

def get_pathway_scores(gsea_dat, alpha = 0.1):
    go_value = []
    random_go_value = []
    gsea_dat = gsea_dat[gsea_dat['BH_p'] < alpha]
    
    for g in set(gsea_dat['group']):
        tp = gsea_dat[gsea_dat['group']==g]
        sig_path_g = np.array(tp.index)

        dat_go_tis = np.intersect1d(Comp_tissues[g], dat.columns)
        if len(dat_go_tis) > 0:
            #print(np.max(dat.loc[np.array([x.replace(' ','') for x in sig_path.index])][dat_go_tis], axis=1).sort_values())
                 
            ## match for number of genes in the pathway
            real_pathways = []
            random_pathways = []
            for ggg in sig_path_g:
                pathway_go_number = np.ceil(len(gsea_pathways[ggg.replace(' ','')])/10)
                pool_g = gsea_pathways_number[pathway_go_number]
                pool_g = set(pool_g) - set([x.replace(' ','') for x in sig_path_g])
            
                if len(pool_g) == 0:
                    pathway_go_number = np.ceil(len(gsea_pathways[ggg.replace(' ','')])/10) + 1
                    try:
                        pool_g = gsea_pathways_number[pathway_go_number]
                        pool_g = set(pool_g) - set([x.replace(' ','') for x in sig_path_g])
                    except:
                        continue
                    
                if len(pool_g) == 0:
                    pathway_go_number = np.ceil(len(gsea_pathways[ggg.replace(' ','')])/10) - 1
                    try:
                        pool_g = gsea_pathways_number[pathway_go_number]
                        pool_g = set(pool_g) - set([x.replace(' ','') for x in sig_path_g])
                    except:
                        continue

                if len(pool_g) == 0:
                    pathway_go_number = np.ceil(len(gsea_pathways[ggg.replace(' ','')])/10) + 2
                    try:
                        pool_g = gsea_pathways_number[pathway_go_number]
                        pool_g = set(pool_g) - set([x.replace(' ','') for x in sig_path_g])
                    except:
                        continue

                if len(pool_g) == 0:
                    pathway_go_number = np.ceil(len(gsea_pathways[ggg.replace(' ','')])/10) - 2
                    try:
                        pool_g = gsea_pathways_number[pathway_go_number]
                        pool_g = set(pool_g) - set([x.replace(' ','') for x in sig_path_g])
                    except:
                        continue

                
                real_pathways.append(ggg.replace(' ',''))
                random_pathways.append(np.random.choice(list(pool_g)))
            
            go_value.append(np.sort(np.max(dat.loc[real_pathways][dat_go_tis], axis=1)))
            random_go_value.append(np.sort(np.max(dat.loc[random_pathways][dat_go_tis], axis=1)))
            #print(g, len(sig_path_g), len(go_value[-1]))
        else:
            print('Not exist', Comp_tissues[g])
        
        
        if 0:
            xxx = tp.loc[sig_path_g]
            xxx =xxx.sort_values('oddsratio')
            sns.set(font_scale=1.5)
            sns.set_style('whitegrid')
            plt.figure(figsize = (8, 0.4 * len(xxx) + 3))
            plt.scatter(xxx['oddsratio'], range(len(xxx)), c=np.array(xxx['BH_p']), cmap='coolwarm_r', s = 100)
            plt.yticks(range(len(xxx)), xxx.index)
            plt.xlabel('Odds ratio')
            plt.colorbar()
            if g == 0:
                main_tissues = 'shared'
            elif g == 1:
                main_tissues = 'brain'
            else:
                main_tissues = ':'.join(Comp_tissues[g])
            plt.title('Enrichment pathway for %s \n at BH-pvalue < 0.1' % main_tissues)
            plt.show()
            plt.close()
    
        if 0:
            plt.figure()
            plt.scatter(go_value[-1], random_go_value[-1])
            plt.plot([0,1], [0,1])
            plt.title(g)
            plt.show()
    
    


    go_value = np.array([a for b in go_value for a in b])
    random_go_value = np.array([a for b in random_go_value for a in b])
    idx = np.where(~np.isnan(go_value))[0]
    
    go_value = go_value[idx]
    random_go_value = random_go_value[idx]
    return [go_value, random_go_value]
```

```python
def qqplot_log10PV(pv_vec, label):
    increment = 1.0 / len(pv_vec)
    x_expect = []
    x_ei = 1.0
    for i in range(len(pv_vec)):
        x_expect.append(x_ei)
        x_ei -= increment
        
    x_expect = np.sort(-np.log10(x_expect))
    pv_vec = np.sort(-np.log10(pv_vec))
    
    if 1:
        plt.scatter(x_expect, pv_vec, label = label)
        xman = np.max([np.max(x_expect), np.max(pv_vec)])
        plt.plot([0, xman], [0,xman])
        
    return [x_expect, pv_vec]
```

```python
### gene TPM 

%run utils.ipynb
[X, Comp_tissues, tissues] = readin_X(FMfn)

#LMfn = 'flashr_Loadings_beta_BH_alpha0.05_corrected'
#[X, Comp_tissues, tissues] = readin_X_flashr()


geneGroup = 'topGene22'
top_genes_group = {}

for k in range(1,23):
    gene_dat = pd.read_csv('%s/%s_%s_group%d.txt' % (pairdir, LMfn, geneGroup, k), sep='\t', header=None)
    geneset = np.array(gene_dat[0])
    for g in geneset:
        if g not in top_genes_group:
            top_genes_group[g] = [k]
        else:
            top_genes_group[g].append(k)
    
```

```python
### gene TPM median


gene_tpm_mt = []
gene_tpm_unmt = []
top_genes = []

for g in np.random.choice(list(top_genes_group.keys()), 1000, replace = False):
    
    top_genes.append(g)
    
    matched_tis = [Comp_tissues[k] for k in top_genes_group[g]]
    matched_tis = np.array([a for b in matched_tis for a in b])
    un_matched_tis = set(list(tissues)) - set(list(matched_tis))
    
    mt_tp = []
    for tis in matched_tis:
        mt_tp.append(np.median(gene_tpm_tissues[tis].loc[g]))
    
    unmt_tp = []
    for tis in un_matched_tis:
        unmt_tp.append(np.median(gene_tpm_tissues[tis].loc[g]))
        
        
    gene_tpm_mt.append(np.median(mt_tp))
    gene_tpm_unmt.append(np.median(unmt_tp))


    
gene_tpm_mt = np.log10(np.array(gene_tpm_mt)+0.001)
gene_tpm_unmt = np.log10(np.array(gene_tpm_unmt) + 0.001)

plt.figure()
plt.hist(gene_tpm_mt, color = 'red', bins = 20, density=True, histtype='step', label = 'matched tissues')
plt.hist(gene_tpm_unmt, color = 'blue', bins = 20, density=True, histtype='step', label = 'unmatched tissues')
plt.legend()
plt.show()
plt.close()

print(ranksums(gene_tpm_mt, gene_tpm_unmt))


dat_tp = pd.DataFrame({"gene": top_genes, "matched": gene_tpm_mt, "un-matched": gene_tpm_unmt})
#dat_tp.to_csv('/work-zfs/abattle4/heyuan/tissue_spec_eQTL_v8/plots/Fig4_eGenes_exp.txt', sep='\t', index = False)

plt.figure()
plt.scatter(dat_tp['un-matched'], dat_tp['matched'], alpha = 0.1)
plt.plot([0,8], [0,8])
plt.show()
plt.close()

```
